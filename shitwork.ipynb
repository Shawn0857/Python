```python
money = int(input("請輸入原始價格"))
if money < 2000 :
  a = money * 0.9
  print("原始價格:{:10.1f},折扣價:{:10.1f}".format(money, a))
elif money < 4000 : 
  b = money * 0.75
  print("原始價格:{:10.1f},折扣價:{:10.1f}".format(money, b))
else:
  c = money * 0.6
  print("原始價格:{:10.1f},折扣價:{:10.1f}".format(money, c))
  
```

```python
#成績分數轉等地
score1= float(input("輸入一個學生分數(0~100):"))
score2= float(input("輸入一個學生分數(0~100):"))

a = ( (score1 * 0.4) + (score2 * 0.6) )
if a<0 or a>100:
  print("輸入範圍錯誤")
else:
  if 90<=a:
    print("A")
  elif 80<=a and a<90:
    print("B")
  elif 70<=a and a<80:
    print("C")
  elif 60<=a and a<70:
    print("D")
  else:
    print("E")
print("總分:{:10.1f}".format(a))
```
