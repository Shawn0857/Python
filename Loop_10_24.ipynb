```python
次數, 目標 = 0, 5
while 次數<目標:
  print("已經練習了{}次\t開始第{}次的練習".format(次數, 次數+1))
  次數 += 1
else:
  print("達成目標,一共練習{}次".format(次數)) 
```

```python
次數 = 1
目標 = int(input("輸入練習次數:"))
while  次數<=目標:
  print("練習第{}次".format(次數))
  次數+=1
else:
  print("達成目標,一共練習{}次".format(次數-1))
```

```python
#for 迴圈

data = ["Mon.","Tuc.","Wcd.","Thr.","Fri."]
for c in data:     #將串列中每個元素依序列出
  print(c,end="")
```

```python
#range函數

seq = range(5)
print(seq)
for c in seq:     
  print(c,end=".")
```

```python
#巢狀式迴圈

data = ["Mon.","Tuc.","Wcd.","Thr.","Fri."]
for c in data:     
  print(c,end="")
  目標 =int(input("今天要練習幾次?"))
  次數=0
  while 次數<目標:
    print("練習第{}次".format(次數+1),end="")
    次數+=1
  else:
    print()     
```
