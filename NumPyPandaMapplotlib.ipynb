```python
import numpy as np #陣列相關套件numerical Python
import pandas as pd #處理一維行別稱Series,處理二維行別稱DataFrame,處理三維行別稱Panel
import matplotlib.pyplot as plt #好用的Python 繪圖套件
array = np.zeros((2,3)) #宣告一個2*3的陣列也就是有兩個一維陣列,每一個一維陣列有三個元素
print(array)
array = np.zeros((2,3),dtype = np.intc)#產生一個內容值皆為0的2*4二維陣列
print(array)
array = np.ones((2,4),dtype = np.intc)#產生一個內容值皆為1的2*4二維陣列
print(array)
array = np.array([[1,2,3],[4,5,6]],dtype=np.intc) #產生一個自訂元素內容2*3二維陣列
print(array)
print("二維陣列的元素大小:",array.size) 
print("二維陣列中第一個陣列中的第二個元素:",array[0][1])
print("二維陣列中第一個陣列所有元素",array[0])
#將二維陣列中第一個陣列中的第二個元素的值改成100
array[0][1]=100
print("二維陣列中第一個陣列所有元素",array[0])
array = np.arange(12) #產生有十二個元素的一維陣列
print(array)
array = np.arange(12,20) #產生一維陣列給定起始值及結束值
print(array)
array = np.arange(12,20,2) #產生一維陣列並給定起始值及結束值,中間間隔為偶數
print(array)
print(array[0]) #陣列第一個值
array[2] = 28 #將陣列中的第三個元素改成28
print(array.size)
#array[5] = 100
#print()
```
