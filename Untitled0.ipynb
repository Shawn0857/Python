```python
imgF = imgFile7
demography = DeepFace.analyze(imgF, actions = ['age', 'gender', 'race', 'emotion'])
reg = demography["region"]
l, t, r, b = reg['x'], reg['y'], reg['x']+reg['w'], reg['y']+reg['h']
img=np.asarray(Image.open(imgF))
cv.rectangle(img, (l, t), (r, b), (255, 0, 0), 5) # 3. 框出人臉
plt.imshow(img)
plt.axis('off')
plt.show()
print("Age: ", demography["age"])
print("Gender: ", demography["gender"])
print("Emotion: ", demography["dominant_emotion"])
print("Race: ", demography["dominant_race"])
```

```python
imgF = imgFile5
facesloc = RetinaFace.detect_faces(imgF)
print("偵測到 ", len(facesloc), " 張臉")

for f in facesloc:
  print(facesloc[f][ 'facial_area'] )
```

```python
img = Image.open(imgFile5) # 1. 讀取資料/轉換型別

for f in facesloc:
 area = facesloc[f]['facial_area']
 xdelta, ydelta = int(min(area[0], img.size[0]-area[2])/2), int(min(area[1], img.size[1]-area[3])/2)
 delta = min(xdelta,ydelta)
 #delta=int(delta*1.5)
 crop_img=img.crop((area[0]-delta, area[1]-delta, area[2]+delta, area[3]+delta))
 plt.imshow(crop_img)
 plt.axis('off')
 plt.show()

 crop_img = np.asarray(crop_img)
 demography = DeepFace.analyze(crop_img, actions = ['age', 'gender', 'race', 'emotion'])
 print()
 print("Age: ", demography["age"])
 print("Gender: ", demography["gender"])
 print("Emotion: ", demography["dominant_emotion"])
 print("Race: ", demography["dominant_race"])
```
