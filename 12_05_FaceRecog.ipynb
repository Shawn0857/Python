```python
import face_recognition, os, warnings
import cv2 as cv
import matplotlib.pyplot as plt
from matplotlib.font_manager  import  FontProperties
from PIL import Image,ImageFilter
import numpy as np
warnings.filterwarnings("ignore")
imgFile1 = "/content/drive/MyDrive/2022Python/image2/AngieChiu_00.jpg"
imgFile2 = "/content/drive/MyDrive/2022Python/image2/Lenna_01.png"
imgFile3 = ""#功課
#12/12號新增
imgFile4 = "/content/drive/MyDrive/2022Python/imagesDB/Audrey_Gregory.jpg"
imgFile5 = "/content/drive/MyDrive/2022Python/imagesDB/humen.jpg"
#font1 = FontProperties()
```

```python
im00 = face_recognition.load_image_file(imgFile2)
face_locs = face_recognition.face_locations(im00)
(top,right,bottom,left) = face_locs[0]
print(left,top,right,bottom)
cv.rectangle(im00,(left,top),(right,bottom),(0,255,0),15)
plt.axis("off")
plt.imshow(im00)
```

```python
#多人辨識
image = face_recognition.load_image_file(imgFile4)
face_locs = face_recognition.face_locations(image)
for (top,right,bottom,left)in face_locs:
  cv.rectangle(image,(left,top),(right,bottom),(255,0,0),4)
plt.axis("off")
plt.imshow(image)
```
