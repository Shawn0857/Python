```python
#輸出字體(這兩行在最上面)
font1=FontProperties(fname="/content/drive/MyDrive/2022Python/楷書體.ttf")
font2=FontProperties(fname="/content/drive/MyDrive/2022Python/隸書體.ttf")
監測時間=list(range(24))#24血壓監測
高血壓=[88,84,80,78,75,70,88,95,112,125,133,140,136,132,128,131,126,120,115,126,120,115,128,119]
低血壓=[66,67,60,67,64,65,61,78,76,60,79,70,71,72,77,66,65,67,68,70,79,77,69,66]
plt.grid()
s1=plt.plot(監測時間,高血壓,c="r",ls=":",marker="D",zorder=2)
s2=plt.plot(監測時間,低血壓,c="b",ls=":",marker="^",zorder=2)
#加上圖例說明
#plt.xlabel("Test_Time")#橫軸說明
#plt.ylabel("Blood_Pressure")#縱軸說明
#plt.title("Day_Pressure")#標題說明
plt.xlabel("橫軸說明",FontProperties=font2,fontsize=14)
plt.ylabel("血壓",FontProperties=font2,fontsize=14)
plt.title("當日血壓變化",FontProperties=font1,fontsize=20)
plt.legend(((s1,s2),"high","low"),loc="best")
plt.grid(zorder=0)
plt.savefig(outimg2,dpi=300)
plt.show()
```
