```python
#城市目的:計算BMI
#作者:謝竣翔
#日期: 2022/10/3

name=input('輸入姓名:')
height=float(input("輸入身高(公尺):"))
weight=float(input("輸入體重(公尺):"))
#計算BMI
bmi=weight/height/height
#格式化輸出
print()
print()
print("--------------------------------")
print("姓名:{}身高:{:8.2f},體重:{:8.2f},bmi為{:10.2f}".format(name,height,weight,bmi))
print("--------------------------------")
```

#格式化輸出

```python
#字串格式
s1,s2='Hello','Python'
print('[0123456789]===[0123456789]')
print('[{:10s}]===[{:>10s}]'.format(s1,s2))#符號 >將字串向右對齊

#整數格式
a,b =-123,456
print('[0123456789]===[0123456789]')
print('[{:<10d}]===[{:>10d}]'.format(a,b))
print('[{:010d}]===[{:010d}]'.format(a,b))#空白位置補0
#浮點數格式
a,b=-12.3,45.6
print('[0123456789]===[0123456789]')
print('[{:<10f}]===[{:>10.3f}]'.format(a,b))
print('[{:010f}]===[{:010.3f}]'.format(a,b))#空白位置補0
```

```python
print(123,456,sep=':')#利用sep這個保留字來指定分隔的符號
print(111,2222)
print('Hello,',end='')#利用sep這個保留字來指定終結的符號
print(' Python')
print('你今天\n好嗎?')#\n這個是換行的意思
print('\\n')#印出\n
print('\\')#印出\
print("111\t 222\t 333\t 444")#使用\t對齊
print("555\t 5522\t 33\t 11144")
s1,s2='Hello','Python'
print(s1,s2)
#格式化輸出字串
print('{},,,,{}'.format(s1,s2))
```
