```python
import random #輸入套件
print(random.random())#產生0<=random()<1的亂數
print(random.randint(10,20))  #randint(a,b)則會產生a<=randint()<=b的亂數
```

```python
print("產生亂數",random.randrange(10))#randint(a)則會產生0<=randint()<a的整數亂數
print("產生亂數",random.randrange(85,100))#randint(a,b)則會產生a<=randint()<b的整數亂數
print("產生亂數",random.randrange(85,100,2))#randint(a,b,s)則會產生間格間隔s,a<=randint()<b的整數亂數
```
