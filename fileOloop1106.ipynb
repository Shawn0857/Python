```python
#串列相關函數

監測時間 = list(range(24)) #24小時血壓監測
高血壓 = [88,84,80,78,75,70,88,84,80,78,75,70,88,84,80,78,75,70,88,84,80,78,75,70]
低血壓 = [69,68,66,67,68,64,69,68,66,67,68,64,69,68,66,67,68,64,69,68,66,67,68,64]
print(min(高血壓),max(高血壓)) #LIST functions
print(len(監測時間),len(高血壓),len(低血壓))
HS,LS = sum(高血壓),sum(低血壓)
print(HS,LS)
print("平均的高血壓{:.1f},平均的低血壓{:.1f}".format(HS/len(高血壓),LS/len(低血壓)))
```
